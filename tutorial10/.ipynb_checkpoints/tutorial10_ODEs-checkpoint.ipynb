```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def ODE1a(y):
    return -y

def analytical1a(t):
    return 2*np.exp(-t)
```

```python
def euler_method(func, h, endpoint, initial_t, initial_y):
    """
    func -- returns ODE
    h    -- stepsize
    endpoint  -- final parameter
    initial_t -- initial parameter
    initial_y -- initial func value
    """
    
    t = [initial_t]
    y = [initial_y]
    
    amount_steps = int((endpoint-initial_t)/h + 0.5)
    for i in range(0,amount_steps):
        t.append(t[-1] + h)
        y.append(y[-1] + h*func(y[-1]))
        
    return np.array(t), np.array(y)
```

```python
t, y = euler_method(ODE1a, 0.1, 20, 0, 2)
```

```python
plt.plot(t, analytical1a(t),ls='dashed',label='analytical')
plt.plot(t, y,label='Euler method')
plt.legend(frameon=False)
plt.show()

plt.title("Difference An-Euler")
plt.plot(t, analytical1a(t)-y,ls='dashed')
plt.show()

```

```python
def rungekutta(func, h, endpoint, initial_t, initial_y):
    t = [initial_t]
    y = [initial_y]
    
    amount_steps = int((endpoint-initial_t)/h + 0.5)
    for i in range(0,amount_steps):
        # Four function evaluations
        k1 = h*func(y[-1])
        k2 = h*func(y[-1]+0.5*k1)
        k3 = h*func(y[-1]+0.5*k2)    
        k4 = h*func(y[-1]+k3)

        y.append(y[-1]+1/6*k1+1/3*k2+1/3*k3+1/6*k4)
        t.append(t[-1]+h)
        
    return np.array(t), np.array(y)
    
    
```

```python
t, y = rungekutta(ODE1a, 0.1, 20, 0, 2)
```

```python
plt.plot(t, analytical1a(t),ls='dashed',label='analytical')
plt.plot(t, y,label='RK method')
plt.legend(frameon=False)
plt.show()

plt.title("Difference An-RK")
plt.plot(t, analytical1a(t)-y,ls='dashed')
plt.show()
```

```python
def modified_midpoint(func, H, n, initial_t, initial_y):
    """
    Advance from a point t to point t+H, by a sequence of n substeps
    of size h = H/n
    """
    h = H/n
    
    # z_(m-2)
    z_m2 = initial_y
    # z_(m-1)
    z_m1 = z_m2 + h*func(z_m2)
    # z_m for m = 2
    z_m = z_m2 + 2*h*func(z_m1)
    
    for m in range(3,n+1): # stop at m=n
        # shuffle all one step forward
        z_m2 = z_m1
        z_m1 = z_m
        z_m = z_m2 + 2*h*func(z_m1)
    
    t = initial_t+H
    y = 0.5*(z_m + z_m1 + h*func(z_m))
    
    return t, y
    
```

```python
t, y = modified_midpoint(ODE1a, 3, 100, 0, 2)

an = analytical1a(t)
print (f"Analytical at        t={t}: {an}")
print (f"Modified midpoint at t={t}: {y}")
```
